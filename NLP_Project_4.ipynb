```python
!pip install pandas scikit-learn

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report



data= pd.read_csv('/content/Fake (1).csv')

train_data, test_data = train_test_split(data, test_size=0.8, random_state=42)


port_vectorizer = TfidfVectorizer(max_features=83000)


train_characteristics = port_vectorizer.fit_transform(train_data['text'])
test_characteristics = port_vectorizer.transform(test_data['text'])

classifier = MultinomialNB()


classifier.fit(train_characteristics, train_data['subject'])
predictions = classifier.predict(test_characteristics)
accuracy = accuracy_score(test_data['subject'], predictions)
classification_rep = classification_report(test_data['subject'], predictions)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_rep)
```
